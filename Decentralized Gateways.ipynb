### What is Essentia? 

![alt text](https://news.bitcoin.com/wp-content/uploads/2018/01/launchpad.png "Logo Title Text 1")

- The decentralized web is fragmented. 
- It uses different protocols i.e IPFS, Ethereum, HTTP, StorJ
- We have to use a chrome extension like MetaMask or different browsers like Mist to access it. 
- Essentia acts as a one-stop gateway to the decentralized web that we can access in the browser
- It gives users access to a Wallet, Storage, an Exchange, and a whole host of other decentralized applications.
- Every Essentia user creates a unique seed which grants access to the whole Essentia framework.
- https://medium.com/essentia_one/essentia-one-launches-11m-fund-to-foster-the-growth-of-projects-built-upon-its-protocol-c3f373fab538
- Essentia implements the following, which we'll learn about: Cross Protocol Bridging,Blockchain seeds, Multi-Signature transfers, Encryption, and Cross Chain Atomic Swaps

![alt text](https://nulltx.com/wp-content/uploads/2018/06/essentia-1.png "Logo Title Text 1")

- Program scripts between all the things integrated.
- Orchestrate operations between centralized and decentralized resources

### Technical Overview

###### 29 Modules, but lets talk about 5

##### ESS-Core 

![alt text](https://image.slidesharecdn.com/bestpracticesforusingsecuringbitcons-alanreiner-140411112918-phpapp01/95/best-practices-for-using-securing-bitcons-alan-reiner-56-638.jpg?cb=1397215899 "Logo Title Text 1")

- This software allows the ESS-ID to destroy / restore / backup the Root-ID
- It also allows the ESS-ID to sign / verify / encrypt / decrypt data
- As well as allow it to generate / destroy / manage the Sub-IDs which have the same exact properties
- multi-signature creation /management capabilities embedded.

###### Demo of Multi-Sig https://gist.github.com/itoonx/53652f9591e9b076663366d99f5afcbd

##### Ess-Base

![alt text](https://cdn-images-1.medium.com/max/761/1*ZUA06XdNnJ7oDh_zLTFm-w.png "Logo Title Text 1")

- An interface to the ESS-CORE API. 
- Only the CORE touches the keys and the seeds. 
- BASE is able to bootstrap the rest of the system by starting with the ESS-Home module and its imprinting data.

##### Ess-Auth

![alt text](https://ai2-s2-public.s3.amazonaws.com/figures/2017-08-08/e23aa56f2e8e209fd29e79570036be4c6a854c8c/3-Figure1-1.png "Logo Title Text 1")

- Manages the creation/editing/deletion of the Authorizations connected with the IDs 
- Also of the controlled or participated (multisign) groups. 
- This module also governs the creation, modification and deletion of groups and their lists of users

##### Ess- Bridge

![alt text](https://pbs.twimg.com/media/CfhoZilW4AAaDCm.jpg "Logo Title Text 1")

- This is an off-chain, hybrid or on-chain modules that connect both decentralized resources and the traditional centralized resources with Essentia. 
- Also enable and allow cross-chain and multi-chain operations with and between different blockchains and protocols. 
- I.e a Dapp or a smart contract to be reached from the “outside” in order to also interact with the traditional resources or input/output systems.

##### Ess-Pay

![alt text](https://icodrops.com/wp-content/uploads/2018/04/Essentia-Token-Distribution.jpg "Logo Title Text 1")

- The module to manage access to encrypted crypto-wallets which can be created, accessed or managed by the Root-ID only. 
- It can be extended with plugins to connect and manage other payment processors and/or wallets. 
- It can also be connected with hardware devices such as Point of Sale solutions through its pluginss

### Atomic Cross Chain Swaps

![alt text](https://nemtech.github.io/_images/guides-transactions-atomic-cross-chain-swap1.png "Logo Title Text 1")

- How to transfer value between blockchains?

Step 1 - Person A's funds are locked on chain
Step 2 - A message will be sent to Person B 
Step 3 - Person B's funds are locked on chain
Step 4 - Person A recieves a message
Step 5 - Off chain, both parties agree on an exchange rate
Step 6 - Transfer is exchanged
